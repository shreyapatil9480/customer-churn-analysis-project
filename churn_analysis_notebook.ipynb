# Customer Churn Analysis

This project provides a comprehensive walkthrough of a synthetic customer churn dataset. The goal is to demonstrate how a business analyst or data analyst can explore data, visualize patterns, and build predictive models to understand customer behavior and retention.

The notebook covers:
- Loading and exploring the synthetic dataset
- Performing exploratory data analysis (EDA) with descriptive statistics and visualizations
- Building predictive models to classify churn using logistic regression and random forest
- Evaluating model performance

The dataset was generated programmatically and is intended for educational purposes. None of the data corresponds to real customers.


```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Set plotting style
sns.set(style='whitegrid')
```

```
# Load the dataset
df = pd.read_csv('/home/oai/share/synthetic_customer_churn.csv')
print('Dataset shape:', df.shape)
df.head()
```

```
# Display basic information about the dataset
print(df.info())

# Summary statistics for numeric columns
df.describe()
```

```
# Visualize distributions of numeric features
numeric_cols = ['age', 'annual_income', 'spending_score', 'membership_duration_months']
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for ax, col in zip(axes.flatten(), numeric_cols):
    sns.histplot(df[col], kde=True, ax=ax, bins=30)
    ax.set_title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Churn rate by gender
plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='churn', data=df)
plt.title('Average Churn Rate by Gender')
plt.show()

# Relationship between spending score, membership duration and churn
plt.figure(figsize=(8,6))
sns.scatterplot(x='spending_score', y='membership_duration_months', hue='churn', data=df, alpha=0.7)
plt.title('Spending Score vs. Membership Duration colored by Churn')
plt.show()
```

```
# Encode categorical variable
df_encoded = df.copy()
le = LabelEncoder()
df_encoded['gender_encoded'] = le.fit_transform(df_encoded['gender'])

# Features and target
X = df_encoded[['age', 'annual_income', 'spending_score', 'membership_duration_months', 'gender_encoded']]
y = df_encoded['churn']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Logistic Regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

print("Logistic Regression Classification Report:
", classification_report(y_test, y_pred_lr))
print("Confusion Matrix:
", confusion_matrix(y_test, y_pred_lr))

# Random Forest model
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Classification Report:
", classification_report(y_test, y_pred_rf))
print("Confusion Matrix:
", confusion_matrix(y_test, y_pred_rf))
```
